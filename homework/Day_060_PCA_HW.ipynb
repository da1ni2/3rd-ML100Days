# 使用手寫辨識資料集, 觀察 PCA 算法

# [作業目標]
- 試著模仿範例寫法, 以變更參數後的 PCA + 邏輯斯迴歸判斷手寫辨識資料集, 觀察不同 component 下正確率的變化

# [作業重點]
- 以手寫辨識資料集, 觀察變更參數後的 PCA 算法, 在取不同 component 時, PCA 解釋度與分類正確率如何變化 (In[5], Out[5])

```python
# 載入套件
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.linear_model import SGDClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")
```

```python
# 定義 PCA 與隨後的羅吉斯迴歸函數
# penalty 改為 L1, max_iter 改為 100
logistic = SGDClassifier(loss='log', penalty='l1', max_iter=100, tol=1e-5, random_state=0)
pca = PCA()
pipe = Pipeline(steps=[('pca', pca), ('logistic', logistic)])

# 載入手寫數字辨識集
digits = datasets.load_digits()
X_digits = digits.data
y_digits = digits.target
```

# 作業
* 將參數 penalty 改為 L1, max_iter 改為 100, 觀察 PCA 與羅吉斯迴歸做手寫辨識的效果

```python
# 先執行 GridSearchCV 跑出最佳參數
param_grid = {'pca__n_components' : [4,10,20,30,40,50,64] , 'logistic__alpha' : np.logspace(-4,4,5)}
search = GridSearchCV(pipe ,param_grid , cv=5 )
search.fit(X_digits , y_digits)
print(f'best score : {search.best_score_}')
print(search.best_params_)
```

```python
# 繪製不同 components 的 PCA 解釋度
pca.fit(X_digits)

fig = plt.figure(figsize = (6,6))
plt.plot(pca.explained_variance_ratio_, linewidth=2)
plt.ylabel('PCA explained variance')

plt.axvline(search.best_estimator_.named_steps['pca'].n_components, linestyle=':', label='n_components chosen')
plt.legend()

plt.show()
```

```python
results = pd.DataFrame(search.cv_results_)
components_col = 'param_pca__n_components'
```

```python
data = results.groupby('param_pca__n_components').apply(lambda g:g.nlargest(1,'mean_test_score'))
```

```python
plt.plot(data['param_pca__n_components'] , data['mean_test_score'])
plt.ylabel('Classification accuracy (val)')
plt.xlabel('n_components')
plt.show()
```
