```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score , mean_squared_error , r2_score
from sklearn.tree import DecisionTreeClassifier , DecisionTreeRegressor
import graphviz
import matplotlib.pyplot as plt
```

```python
data = datasets.load_iris()
```

```python
x_train , x_test , y_train , y_test = train_test_split(data.data , data.target , test_size = 0.25 , random_state = 4)
```

```python
cls = DecisionTreeClassifier(criterion = 'entropy' , random_state = 1)
```

```python
cls.fit(x_train ,y_train)
y_pred = cls.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
